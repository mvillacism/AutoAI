```python
# Carga de Librerías 
import numpy as np
import pandas as pd
import os
```

```python
# Carga de Datos
path = 'G:/respaldos jc/Archivos_Gidtec/Proyecto Aceleración AutoML España/Revisión para prueba de concepto/Datos Open Wind/'
dataName = 'wind-farm-1-signals-2016.csv'
data = pd.read_csv(path + dataName,sep=';')
```

```python
# Verificar los primeros datos
data.head()
```

```python
# Convertimos la columna de Timestamp en un objeto datetime
data["Timestamp"] = pd.to_datetime(data["Timestamp"])
data.head()
```

```python
# Generar los archivos .hdf5
save_Path = 'G:/respaldos jc/Archivos_Gidtec/Proyecto Aceleración AutoML España/Base de Datos' # Poner el nombre de la carpeta que lo va a contener
if os.path.exists(path) == False:
    os.makedirs(save_Path, exist_ok=False)
    print("The new directory is created!")
Turbines = data["Turbine_ID"].unique()
for turbine in Turbines:
    isExist = os.path.exists(save_Path+'/'+turbine)
    if not isExist:
        os.makedirs(save_Path+'/'+turbine, exist_ok=False)
        print("The new directory is created!")
    save_Data = data[data['Turbine_ID'] == turbine].sort_values(by='Timestamp')
    save_Data.to_hdf(save_Path+ '/' + turbine +  '/' + turbine + '.h5' , key='df', mode='w') 
    
    
```

```python
# Cargamos el archivo de log de fallas y ajustamos el formato del Timestamp
failureName = 'htw-failures-2016.csv'
failures = pd.read_csv(path + failureName,sep=';')
failures["Timestamp"] = pd.to_datetime(failures["Timestamp"])
failures = failures.loc[:, ~failures.columns.str.contains('^Unnamed')]
failures.head()
```

```python
# Guardamos un archivo log de fallos con extensión .hdf
failures.to_hdf(save_Path+ '/' + 'failures' + '.h5' , key='df', mode='w') 
```
